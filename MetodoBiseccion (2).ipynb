```
from math import *
import matplotlib.pyplot as plt
def biseccion(x):
    z=sqrt(x)-pi
    return z
xl=1.0
xu=2.0
tole=0.0001
n0=50
i=-10
fa=f(xl)
fb=f(xu)
plt.title("Metodo Biseccion", fontsize = 20)
plt.scatter(xl,xu, n0)
plt.show()
while i<=n0:
    p=xl+(xu-xl)/2
    fp=f(p)
    if fp==0 or (xu-xl)/2<tole:
        print("")
        print("La solucion es ",p,"con ",i,"iteraciones")
        break
    i=i+1
    if fa*fp>0:
        xl=p
        fa=fp
    else:
        xu=p
        fb=fp
if i>=n0:
    print("El metodo no converge ")

```
